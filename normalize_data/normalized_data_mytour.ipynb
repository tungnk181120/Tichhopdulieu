# **Kiểm tra các thuộc tính Missing-value khác**

```python
percent_missing = raw_mytour.isnull().sum() * 100 / len(raw_mytour)
missing_value_df = pd.DataFrame({'column_name': raw_mytour.columns,
                                 'percent_missing': percent_missing})
```

```python
missing_value_df
```

```python
#Bởi vì không thể thêm các giá trị Min Price cho bản ghi bị missing-value, do đó ta sẽ xóa hết các bản ghi này
raw_mytour = raw_mytour.dropna(subset=["Min Price"])
```

```python
raw_mytour = raw_mytour.reset_index(drop=True)
```

```python
raw_mytour.info()
```

Nhận thấy, sau khi loại bỏ các bản ghi missing-value về giá; thì đã không còn bản ghi nào bị thiếu nữa.

# **Chuẩn hóa dữ liệu cột Min Price**

```python
raw_mytour['Min Price'][3]
```

```python
#Bỏ "đ"
raw_mytour["Min Price"] = raw_mytour["Min Price"].apply(lambda x: x.replace('đ',''))
```

```python
#Chuyển về kiểu int và bỏ dấu "." giữa số
raw_mytour["Min Price"] = raw_mytour["Min Price"].apply(lambda x: int(x.replace('.','')))
```
