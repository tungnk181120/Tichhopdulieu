```python
import pandas as pd
import re
```

```python
booking = df = pd.read_csv('../raw_data/craw_booking_com_full.csv')
booking.info()
```

```python
booking.head()
```

```python
booking['Hotel name'] = booking['Hotel name'].apply(lambda x: x.replace('\n', ' '))
booking.head()
```

```python
booking['Rating'] = booking['Rating'].apply(lambda x: float(str(x).replace(',', '.')))
booking.head()
```

```python
booking['Price'] = booking['Price'].apply(lambda x: int(x.replace('VND', '').replace('.', '')))
booking.head()
```

```python
booking['Number of reviews'] = booking['Number of reviews'].apply(lambda x: re.sub('[^0-9]','', str(x)))
booking.head()
```

```python
booking[(booking['Number of reviews'] == '')] = 1
booking.head()
```

```python
booking['Number of reviews'] = booking['Number of reviews'].apply(lambda x: int(x))
booking.head()
```

```python
booking.info()
```

```python
booking.to_csv('booking_normalize.csv', index=False, encoding='utf8')
```
