# **Lọc bỏ missing-value**

Do có những list bị rỗng --> Thay list rỗng bằng NULL value

```python
#Thay '[]' bằng NULL
traveloka_df["Image"] = traveloka_df["Image"].apply(lambda y: np.nan if len(y)==2 else y)
```

```python
traveloka_df.info()
```

```python
percent_missing = traveloka_df.isnull().sum() * 100 / len(traveloka_df)
missing_value_df = pd.DataFrame({'column_name': traveloka_df.columns,
                                 'percent_missing': percent_missing})
```

```python
missing_value_df
```

Như vậy, có thể loại bỏ những bản ghi bị thiếu địa chỉ (chỉ chiếm 0.6-0.8% dữ liệu)

```python
#Loại bỏ missing-value records
traveloka_df = traveloka_df.dropna()
```

```python
traveloka_df.info()
```

# **Chuẩn hóa dữ liệu cột Min Price**

```python
traveloka_df = traveloka_df.reset_index(drop=True)
```

```python
# traveloka_df["Hotel name"][100]
```

```python
# demo_df["Hotel name"] = demo_df["Hotel name"].apply(lambda x: x.replace('\n', ' '))
# demo_df.head()
```

```python
#Bỏ "VND"
traveloka_df["Min Price"] = traveloka_df["Min Price"].apply(lambda x: x.replace('VND',''))
```

```python
#Chuyển về kiểu int và bỏ dấu "." giữa số
traveloka_df["Min Price"] = traveloka_df["Min Price"].apply(lambda x: int(x.replace('.','')))
```
